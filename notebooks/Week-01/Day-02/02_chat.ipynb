```python
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv()
```

```python
client = OpenAI()
```

**Zero Shot Prompting**

```python
response = client.chat.completions.create(
    model="gpt-4", messages=[
        {"role": "user", "content": "Hey there"}
    ]
)

print(response.choices[0].message.content)
```
