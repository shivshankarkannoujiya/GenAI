**Few Shot Prompting**

```python
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv()
```

```python
client = OpenAI()
```

```python
system_prompt = """
You are an AI Assistent who is specialized in Maths.
You should answer the the mathmatics questions asked by user in easy language.
You should not answer that do not belongs to the maths.

for a given query help user to solve that along with explanation

EXAMPLE:

input: 2 + 2
output: 2 + 2 is 4 which is calculated by adding 2 with 2

input: 3 * 10
output: 3 * 10 is 30 which is calculated by multiplying 3 wit 10, 
Funfact you can even multiply 10 with 3 which will give the same result

input: why sky s=is Blue
output: Hey man You alright ? is it math query 
"""
```

```python
response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": "what is 2 + 3 ?"}
    ]
)

print(response.choices[0].message.content)
```
