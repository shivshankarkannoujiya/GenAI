```python
import tiktoken # OpenAI Library
```

```python
encoder = tiktoken.encoding_for_model("gpt-4o")
```

**VOCAB SIZE: UNIQUE TOKENS**

```python
print("Vocab size: ", encoder.n_vocab)  # 200019: GPT-4o
```

```python
text = "The cat sat on the mat"
tokens = encoder.encode(text)
print("TOKENS: ", tokens)
```

```python
my_tokens = [976, 9059, 10139, 402, 290, 2450]
deocdedText = encoder.decode(my_tokens)
print("DECODED: ", deocdedText)
```
